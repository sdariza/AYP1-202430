Para anidar condicionales no repetitivos en python, hacemos uso de la palabra `elif`

Sin elif

```python
#INICIO
if False:
    print("Hola")
elif True:
    print("Mundo")
else:
    print("Python")
#FIN
```

Con elif

```python

#INICIO
if False:
    print("Hola")
else:
  if True:
    print("Mundo")
  else:
    print("Python")
#FIN
```

La siguiente solución es sin elif

```python
#INICIO
valor_imp = 0
num_copias = int(input("Ingrese la cantidad de copias a sacar: "))
valor_imp = num_copias * 300
print(f"El valor a pagar sin impuesto es: {valor_imp}")
if valor_imp >=3000 and valor_imp <=10000:
  valor_imp = valor_imp * (1 + 0.02)
else:
  if valor_imp >=10001 and valor_imp <=20000:
    valor_imp = valor_imp * (1 + 0.05)
  else:
    if valor_imp >=20001 and valor_imp <=30000:
      valor_imp = valor_imp * (1 + 0.1)
    else:
      if valor_imp >30001:
        valor_imp = valor_imp * (1 + 0.15)
print(f"El valor a pagar con impuesto es: {valor_imp}")
#FIN
```

Usando elif

```python
#INICIO
valor_imp = 0
num_copias = int(input("Ingrese la cantidad de copias a sacar: "))
valor_imp = num_copias * 300
print(f"El valor a pagar sin impuesto es: {valor_imp}")
if valor_imp >=3000 and valor_imp <=10000:
  valor_imp = valor_imp * (1 + 0.02)
elif valor_imp >=10001 and valor_imp <=20000:
  valor_imp = valor_imp * (1 + 0.05)
elif valor_imp >=20001 and valor_imp <=30000:
  valor_imp = valor_imp * (1 + 0.1)
elif valor_imp >30001:
  valor_imp = valor_imp * (1 + 0.15)
print(f"El valor a pagar con impuesto es: {valor_imp}")
#FIN
```
