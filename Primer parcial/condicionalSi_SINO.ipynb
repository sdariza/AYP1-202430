# Realice un algoritmo para leer dos números y determine cuál de los dos es mayor.

```python
#INICIO
n1 = float(input("Ingrese el primer número: "))# escribir y leer el primer número
n2 = float(input("Ingrese el segundo número: "))# escribir y leer el 2ndo número
print("n1=",n1,"n2=",n2)#escribimos los números leídos
print(f"n1={n1} y n2= {n2}")
if n1 < n2: # si el primer número es menor que el segundo
  print(f"{n2} es mayor")# el segundo es el mayor
else:# si no, quiere decir que el primero es el mayor
  print(f"{n1} es mayor")# el primero es el mayor
#FIN
```
