Realice un algorItmo en python para calcular el promedio de 4 notas de un estudiante.



```
INICIO
|    REAL: nota1, nota2, nota3, nota4, promedia<-- 0
|    ESC("Ingrese la nota 1")
|    lea nota1
|    ESC("Ingrese la nota 2")
|    lea nota2
|    ESC("Ingrese la nota 3")
|    lea nota3
|    ESC("Ingrese la nota 4")
|    lea nota4
|    promedio <-- (nota1+nota2+nota3+nota4)/4
|    ESC("El promedio calculado es: ", promedio)
FIN
```



````python
#INICIO
nota1 = float(input("Ingrese la nota 1 "))
nota2 = float(input("Ingrese la nota 2 "))
nota3 = float(input("Ingrese la nota 3 "))
nota4 = float(input("Ingrese la nota 4 "))
promedio = (nota1+nota2+nota3+nota4)/4
print("El promedio calculado es: ", promedio)
#FIN
````

Realice un algoritmo para leer el nombre, la edad, el número de horas trabajadas, el número de horas extras y el valor de la hora de un empleado. Asimismo, calcule cuánto se le debe pagar, si la hora extra tiene un recargo de un 20% de la hora normal.

````python
#INICIO
nombre = input("Ingrese el nombre del empleado ") #input retorna lo leído en str
edad = int(input("Ingrese la edad del empleado "))
horas_trabajadas = float(input("Ingrese el número de horas trabajadas "))
horas_extras = float(input("Ingrese el número de horas extras "))
valor_hora = float(input("Ingrese el valor de la hora "))
salario_basico = horas_trabajadas  *valor_hora
ganancia_extra = (valor_hora*0.2+valor_hora) * horas_extras
salario = salario_basico + ganancia_extra
print("El salario del empleado ", nombre, "es: ", salario)
#FIN
````
