````python
    #INICIO
N = int(input("Ingrese la cantidad de nmeros: "))
i = 1
contPos = 0
contNeg = 0
contCeros = 0
sumNums = 0
while i <= N:
  num = int(input("Ingrese el nmero: "))
  if num > 0:
    contPos = contPos + 1
    sumNums = sumNums + num
  #FIN-SI
  if num < 0:
    contNeg = contNeg + 1
  #FIN-SI
  if num == 0:
    contCeros = contCeros + 1
  #FIN-SI
  i = i + 1
#FIN-MQ
promedioPos = sumNums / contPos
print(f"La cantidad de nmeros positivos es: {contPos}")
print(f"La cantidad de nmeros negativos es: {contNeg}")
print(f"La cantidad de ceros es: {contCeros}")
print(f"El promedio de nmeros positivos es: {round(promedioPos,2)}")
#Tarea: pasar al cdigo el pseudo realizado.
````
