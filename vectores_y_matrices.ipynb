Recuerde que esta estructura es lineal, por tanto la forma de accederla es de forma secuencial. Es decir se recorre desde el primer elementos (pos 0), hasta la último elemento (pos n-1) donde n es la cantidad de elementos en el vector.

````python
vector_nums = [251, 56, 12, 58, 1101, 1001, 4005, 10, 1001, 554]
n = len(vector_nums) # calcular el número de elementos
for sub_indice in range(0,n,1):
  print(f"vector_nums[{sub_indice}] = {vector_nums[sub_indice]}")
````

# **Matrices**
Se les conoce como una estructura bidimensional, están basadas en las matrices de álgebra lineal. Con la salvedad que las matrices pueden almacenar cadenas u otro tipo de dato no número. Sin embargo, a nivel del curso. Las matrices solo almancenan valores numéricos.

Asimismo, este tipo de estructura la explicaremos basados en las listas. Por ende, a una matriz la conoceremos una lista de lista.

````python
A = [[5, 20, 30], [50, 20, 3], [15, 21, 13], [55, 26, 31]]
print(A[2][1])
````

Observación: note que es neceseario iterar sobre las filas y las columnas. Por lo tanto es necesario tener 2 subíndices, que por lo general son i y j.

````python
for i in range(0,4,1):
  print(i, end=" ")
  for j in range(0,3,1):
    print(j, end=" ")
  print()
````

Para recorrer la matriz de forma secuncial, debemos tener 2 ciclos for, el primero controla las filas a recorrer, y el interno (segundo) controla las columnas recorridos para la fila que se recorre.

````python
for i in range(0,4,1):# recorre las filas
  for j in range(0,3,1): # recorro las columnas de cada fila
    print(f"{A[i][j]}", end="|")
  # se acabaron las columna de la fila, por ende avanzo a la sgte fila
  print("")
````

````python
# Por lo pronto, supongamos que todas las sucursales venden 4 artículos.
sucursales = [[10, 20, 30, 40], [21, 20, 60, 50], [0, 1, 2, 3], [10, 30, 20, 50], [89, 50, 21, 10]]
precios = [20000, 30000, 10000, 50000]
ganancias = [0, 0, 0, 0, 0]
````

````python
for i in range(0,5,1):# recorre las filas
  for j in range(0,4,1): # recorro las columnas de cada fila
    print(f"{sucursales[i][j]}", end="|")
  # se acabaron las columna de la fila, por ende avanzo a la sgte fila
  print("")
````

# **Punto 1**

Calculamos en un vector las ganancias de cada sucursal

````python
for i in range(0,5,1):# recorre las filas
  sum = 0
  for j in range(0,4,1): # recorro las columnas de cada fila
    sum = sum + sucursales[i][j] * precios[j]
  ganancias[i] = sum
  # se acabaron las columna de la fila, por ende avanzo a la sgte fila
print(ganancias)
````

Calculamos la mayor ganacia

````python
mayor = ganancias[0]
for i in range(1, 5, 1):
  if mayor < ganancias[i]:
    mayor = ganancias[i]
````

Recorremos las ganancias de nuevo e indicamos qué sucursales tienen el valor de la ganancia igual a la de mayor de ganacia.

````python
for i in range(0, 5, 1):
  if mayor == ganancias[i]:
    print(f"La sucursal {i + 1} tiene la mayor ganancia")
````
